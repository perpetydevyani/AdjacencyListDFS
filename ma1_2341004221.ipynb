```python
## Assignment Overview

#The program calculates:
#1. Basic salary based on hourly rate and weekly hours.
#2. Overtime salary for hours above 40 per week.
#3. Tax deductions based on salary slabs.
#4. Gross salary after allowances and tax.
#5. Salary bracket classification.
#6. A report for three employees.

#All formulas strictly follow the assignment guidelines.
```

```python
def compute_basic(rate, weekly_hours):
    """Returns monthly basic salary assuming 4 weeks."""
    normal_hours = min(weekly_hours, 40)
    return rate * normal_hours * 4


def compute_overtime(rate, weekly_hours):
    """Returns monthly overtime salary."""
    if weekly_hours > 40:
        extra = weekly_hours - 40
        return extra * rate * 1.5 * 4
    return 0


def compute_total(rate, weekly_hours):
    """Returns total salary (basic + overtime)."""
    base = compute_basic(rate, weekly_hours)
    extra = compute_overtime(rate, weekly_hours)
    return base + extra


def tax_deduction(basic_amount):
    if basic_amount < 60000:
        return 0.10 * basic_amount
    elif 60000 <= basic_amount <= 85000:
        return 0.15 * basic_amount
    else:
        return 0.20 * basic_amount


def compute_gross(basic_amount):
    allowances = 0.20 * basic_amount
    tax = tax_deduction(basic_amount)
    return basic_amount + allowances - tax


def bracket_label(gross_amount):
    if gross_amount < 50000:
        return "Low income"
    elif gross_amount <= 80000:
        return "Middle income"
    else:
        return "High income"


def employee_report():
    print("\n======= EMPLOYEE SALARY REPORT =======\n")

    employees = []

    for idx in range(3):
        print(f"Enter details for Employee {idx + 1}:")
        name = input("Name: ")
        rate = float(input("Hourly Rate: "))
        hours = float(input("Hours per Week: "))

        employees.append((name, rate, hours))
        print()

    for name, rate, hours in employees:
        basic = compute_basic(rate, hours)
        gross = compute_gross(basic)
        tax = tax_deduction(basic)
        category = bracket_label(gross)

        print(f"Employee: {name}")
        print(f"  Basic Salary   : Rs. {basic:.2f}")
        print(f"  Gross Salary   : Rs. {gross:.2f}")
        print(f"  Tax Deducted   : Rs. {tax:.2f}")
        print(f"  Salary Bracket : {category}")
        print("-------------------------------------")
```

```python
employee_report()
```
